# Which countries have never won a gold medal but have silver or bronze?

```python
data
```

```python
df = data.loc[data["Gold"]==0 & ((data["Silver"]>0 )| (data["Bronze"]>0))]
df
```

```python
df.to_csv("countries_with_no_gold_medal_but_has_silver_and_bronze.csv", index=False)
```

# Which countries have the same number of gold, silver, and bronze medals?

```python
df1 = data.loc[(data["Gold"]==data["Silver"]) &(data["Silver"]==data["Bronze"])]
df1
```

# Which countries have won more silver than gold medals?

```python
df2 = data.where(data["Gold"]<data["Silver"]).dropna()
df2
```

# What is the average number of medals won (gold/silver/bronze)?

```python
data
```

```python
average_gold = data["Gold"].mean()
average_silver = data["Silver"].mean()
average_bronze = data["Bronze"].mean()
```

```python
print("the average number of gold medal won by the countries is " , average_gold)
print("the average number of silver medal won by the countries is " , average_silver)
print("the average number of bronze medal won by the countries is " , average_bronze)
```

# Which country has the highest total to gold medal ratio?

```python
data
```

```python
data["Gold Medal Ratio"] = data["Gold"] / data["Total"]
data
```

```python
data.to_csv("countries_with_gold_medal_ratio.csv", index=False)
```

```python
data
```

```python
max = (data["Gold Medal Ratio"].max())
```

```python
filtered = data.where(data["Gold Medal Ratio"] == max).dropna()
```

```python
filtered.to_csv("countries_with_max_gold_medal_ratio.csv", index=False)
```
